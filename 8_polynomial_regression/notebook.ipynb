# Polynomial Regression

# Importing libraries and datset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/Position_Salaries.csv")
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
X
```

```python
y
```

# Training the model

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```


# Visualising the plots for linear_regression 1

```python
plt.scatter(X,y,color ='red')
plt.plot(X,lin_reg.predict(X))
plt.title("Is He Lying?")
```


# Visualising the plots for polynomial_regression

```python
plt.scatter(X,y,color ='red')
plt.plot(X,lin_reg_2.predict(X_poly))
plt.title("Is He Lying?")
```

# for degree 3

```python
poly_reg_3 = PolynomialFeatures(degree = 3)
X_poly_3 = poly_reg_3.fit_transform(X)
lin_reg_3 = LinearRegression()
lin_reg_3.fit(X_poly_3,y)

plt.scatter(X,y,color ='red')
plt.plot(X,lin_reg_3.predict(X_poly_3))
plt.title("Is He Lying?")
```

```python
poly_reg_4 = PolynomialFeatures(degree = 4)
X_poly_4 = poly_reg_4.fit_transform(X)
lin_reg_4 = LinearRegression()
lin_reg_4.fit(X_poly_4,y)

plt.scatter(X,y,color ='red')
plt.plot(X,lin_reg_4.predict(X_poly_4))
plt.title("Is He Lying?")
```

```python
poly_reg_5 = PolynomialFeatures(degree = 5)
X_poly_5 = poly_reg_5.fit_transform(X)
lin_reg_5 = LinearRegression()
lin_reg_5.fit(X_poly_5,y)

plt.scatter(X,y,color ='red')
plt.plot(X,lin_reg_5.predict(X_poly_5))
plt.title("Is He Lying?")
```
